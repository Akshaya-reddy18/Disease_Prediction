```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
import pickle
```

```python
pcod_dataset = pd.read_csv("../datasets/pcod.csv")
```

```python
pcod_dataset = pcod_dataset.dropna(subset=['Outcome'])
```

```python
pcod_dataset.head()
```

```python
pcod_dataset.describe()
```

```python
pcod_dataset['Outcome'].value_counts()
```

```python
x = pcod_dataset.drop(columns=['Outcome'], axis=1)
y = pcod_dataset['Outcome']
```

```python
x.isna().sum()
y.isna().sum()
```

```python
x = x.fillna(x.mean())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
x_train_imputed = imputer.fit_transform(x_train)
x_test_imputed = imputer.transform(x_test)
```

```python
print(x_train.shape, x_test.shape)
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train, y_train)
```

```python

y_pred = model.predict(x_test_imputed)
```
